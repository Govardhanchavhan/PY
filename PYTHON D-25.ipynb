```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

#Line Plot

```python
plt.plot(titanic.index, titanic['age'], color = 'pink', linestyle = '--', marker = 'x')
plt.title('Passenger Index vs Age')
plt.xlabel('Passenger Index')
plt.ylabel('Age')
plt.grid()
plt.show()
```

#Scatter Plot

```python
plt.scatter(titanic['age'], titanic['fare'], c = 'red')
plt.title('age vs fare')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

#Bar Plot

```python
survival_counts = titanic['survived'].value_counts()
```

```python
survival_counts
```

```python
plt.bar(['Not Survived', 'Survived'], survival_counts)
plt.title('survival count')
plt.xlabel('survival status')
plt.ylabel('counts')
plt.show()
```

# Histogram Plot

```python
gender_count = titanic['sex'].value_counts()
```

```python
gender_count
```

```python
plt.hist(titanic['age'].dropna(), bins = 20, color= 'purple', edgecolor = 'black')
plt.title('age histogram')
plt.xlabel('age')
plt.ylabel('freq.')
plt.show()
```

#Pi Chart

```python
survived = titanic['survived'].value_counts()
```

```python
survived
```

```python
labels = ['not survived', survived]
colors = ['pink', 'yellow']
```

```python
plt.pie(survived, labels = labels,autopct = '%1.1f%%', colors = colors)
```
