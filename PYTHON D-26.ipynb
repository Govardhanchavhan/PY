```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

#Scatter plot

```python
sns.scatterplot (x = 'age', y= 'fare', data = titanic, hue = 'who', palette= 'magma')
plt.title('Scatter plot')
```

#Line Graph

```python
sns.lineplot(x= 'age', y= 'fare',data= titanic)
```

```python
sns.lineplot(x= 'age', y= 'fare',data= titanic, hue = 'class', palette= 'coolwarm', linewidth= 3)
```

#Bar Graph

```python
sns.barplot (x='class', y = 'fare', data= titanic,hue = 'sex', palette= 'pastel')
```

#Count Plot

```python
sns.countplot(x = 'class', data = titanic, hue = 'sex', dodge= True)
```

#Box Plot

```python
sns.boxplot(x = 'class', y = 'fare', data = titanic)
```

```python
sns.boxplot(x = 'class', y = 'fare', data = titanic, hue = 'class')
```

#Violin Plot

```python
sns.violinplot(x= 'class', y = 'fare', data = titanic)
```

#Histogram Plot

```python
sns.distplot (titanic['age'])      #distance plot (no x and y consider)
```

```python
sns.kdeplot(titanic['age'], shade = True)
```

#Pair Plot

```python
sns.pairplot(titanic)
```
